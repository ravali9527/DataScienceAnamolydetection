```python
DIR='/Users/ravali/Desktop/DataScienceAnamolydetection/Data'
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
df1 = pd.read_csv(DIR+'/speed.csv', delimiter=',')
print (df1)
print(df1.dtypes)
```

```python
obj_df = df1.select_dtypes(include=['object']).copy()
obj_df.head()
obj_df["timestamp"] = obj_df["timestamp"].astype('category')
obj_df["timestamp"] = obj_df["timestamp"].cat.codes
display(obj_df)
```

```python
df_final = pd.concat([df1, obj_df], axis=1)
display(df_final)
```

```python
X=df_final.drop('timestamp',axis=1)
display(X)
```

```python
Y=X.value #output label
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
#print(X_train)
print(y_train)
```

```python
from sklearn.cluster import DBSCAN
outlier_detection = DBSCAN(
  eps = 0.5,
  metric="euclidean",
  min_samples = 3,
  n_jobs = -1)
clusters = outlier_detection.fit_predict(X_test)
display(clusters)#the values with -1 are outliers
```
